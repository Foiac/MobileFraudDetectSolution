# Data Job Clean

#### Import dependecies

```python
from pyspark.sql.functions import col, cast, explode, from_json
from pyspark.sql import functions as F
from pyspark.sql.types import *
import os
```

#### Path Definitions

```python
storage_account_name  = os.getenv("ENV_STORAGE")

database_name_input = "mobile"
table_name_input = "access"
delta_table_path_input = f"abfss://contfraud@{storage_account_name}.dfs.core.windows.net/Bronze/{database_name_input}/{table_name_input}"

database_name_output = "fraud"
table_name_output = "tab_mobil_access"
delta_table_path_output = f"abfss://contfraud@{storage_account_name}.dfs.core.windows.net/Silver/{database_name_output}/{table_name_output}"

storage_account_key = os.getenv("ENV_ADLS_KEY")
```

#### ADLS Configuration

```python
spark.conf.set(f"fs.azure.account.key.{storage_account_name}.dfs.core.windows.net", storage_account_key)
```

#### Database Create

```python
spark.sql(f"CREATE DATABASE IF NOT EXISTS {database_name_output}")
```

#### Read Databricks Catalog Table

```python
df = spark.table(f"{database_name_input}.{table_name_input}")
display(df)
```

#### Remover linhas duplicadas

#### Remover colunas nulas ou vazias

```python
#Filtrando linhas onde a coluna timestamp está nulo ou vazio
df_filtered = df.filter(F.col("timestamp").isNotNull() & (F.col("timestamp") != ''))
```

#### Conversão de tipos

```python
df = (df
      .withColumn("transaction", F.when(F.col("transaction") == "true", True).when(F.col("transaction") == "false", False).otherwise(None)
      )
    )
```

#### Renomear colunas

```python
df_silver = (df
             .withColumnRenamed("imei", "IMEI")
             .withColumnRenamed("mac", "MAC")
             .withColumnRenamed("rede", "REDE")
             .withColumnRenamed("client_ip", "IP")
             .withColumnRenamed("latitude", "LATITUDE")
             .withColumnRenamed("logintude", "LONGITUDE")
             .withColumnRenamed("cpf", "UID")
             .withColumnRenamed("senha", "PASSWORD")
             .withColumnRenamed("transaction", "TRANSACTION")
             .withColumnRenamed("api", "API")
             .withColumnRenamed("endpoint", "ENDPOINT")
             .withColumnRenamed("os", "OPERAT_SYSTEM")
             .withColumnRenamed("os_version", "OS_VERSION")
             .withColumnRenamed("app_version", "APP_VERSION")
             .withColumnRenamed("erro", "CD_ERRO")
             .withColumn("DT_EVENT", F.from_unixtime(F.col("timestamp").cast("long")/1000, "yyyy-MM-dd HH:mm:ss").cast("timestamp"))
             .withColumnRenamed("timestamp", "EVENT_TIMESTAMP")
             )
display(df_silver)
```

#### Validação de Dados
- Regra de negócio 1
- Regra de negócio 2
- Regra de negócio 3

#### Escrever dados em tabela silver

```python
df_silver.write\
    .format("delta")\
    .mode("append")\
    .option("path", delta_table_path_output)\
    .saveAsTable(f"{database_name_output}.{table_name_output}")
```
